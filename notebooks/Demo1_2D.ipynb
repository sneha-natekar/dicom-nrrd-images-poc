### Installation of the pydicom library
##### Pydicom is an open-source library for working with DICOM (Digital Imaging and Communications in Medicine) medical image files in Python. DICOM is the standard format used for medical images, such as X-rays, CT scans, MRIs, and ultrasound images. Pydicom is designed to be easy to use and flexible, with support for a wide range of DICOM file formats and data types. It also includes a number of utility functions and classes for common tasks, such as sorting and grouping DICOM files, extracting metadata, and converting between DICOM and other image formats.

#### Reading the file in form of DCM format by providing specific path of file.

```python
# Loading python libraries
import pydicom
import matplotlib.pyplot as plt
# Load the DICOM file and store it in the variable
ds = pydicom.dcmread('/workspaces/codespaces-jupyter/data/0002.DCM')
print(ds)
```

### Extracting the array content of the images from the DCM image objects.

```python
pixel_array = ds.pixel_array
```

### By using the shape element, we can conclude the number of images are present in single image file.

```python
pixel_array.shape
```

### Plotting the images in a grid format from a pixel_array and try to visulize it.

```python
plt.figure(figsize=(15,15)) # specifying the overall grid size
### Plotting images one by one
for i in range(17):
    plt.subplot(5,4,i+1)    # the number of images in the grid is 4*5 (20)
    plt.imshow(pixel_array[i,:,:])

plt.show()
```
